```python
#THIS RETURNS YAHOO QUESTIONS IN CSV AS STRING MAKE A CLASS

import lxml, re
from lxml import etree
from lxml.html.clean import Cleaner
from bs4 import BeautifulSoup as b
import urllib
import urllib.request

#url="https://answers.yahoo.com/dir/index?sid=396545660"
url = "https://answers.yahoo.com"
page = urllib.request.urlopen(url).read()
html = b(page,'lxml')
##FINDS ALL QUESTIONS (MAX 20) ON PAGE AND PRINTS
links = html.findAll("a")

postcontainers = html.findAll("li",{"class":"ya-discover-tile ya-discover-tile-qn Bfc P-14 Bdbx-1g Bgc-w"})
numq=[]
numq = len(postcontainers)
questions =[]
qwlink = []
for i in range(0,numq-1,):
    qcont2 = postcontainers[i]
    var = qcont2.findAll("a",{"class":"Fz-14 Fw-b Clr-b Wow-bw title"})
    newvar = str(var[0])
    newvar2 = re.sub('<.*?>','',newvar)
    #print(newvar2)
    questions.append(newvar2)
    qwlink.append(var)
#print(questions)
#print(qwlink)

qcont3 = html.findAll("div",{"class":"Bfc"})
print(qcont3)

#TEMPLATE TO GET LINKS TO QUESTIONS:
#qcont3 = html.findAll("div",{"class":"Clr-888 Fz-12 Lh-18"})
#num2 = len(qcont3)
#print(qcont3)
#justlinks=[]
#for anscont in qcont3:
    #justlinks = anscont.a["href"]
#    justlinks.append(anscont.a["href"])
#print(justlinks)
#print(len(justlinks))
#newurl=[]
#for i in justlinks:
#    newurl.append("https://answers.yahoo.com"+i)
#print(newurl)


filename = "questions.csv"
f = open(filename,"w")
headers = "Yahoo Questions\n"
f.write(headers)
for i in questions:
    f.write('%s\n'% i)
f.close()


```
